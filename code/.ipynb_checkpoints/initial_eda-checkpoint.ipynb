### Feature descriptions:
- **start_dow** : an oridinal value for the day of the week, changed from the original text to an integer, where 0=Monday...6=Sunday (this is the default convention of pandas)
- **engage** : number of days that the user engaged (unlocked a deal) in the app.
- **gap**: number of days that the user had no activity in the app.
- **session** : number of days that the user went into the app, but did not unlock any deals.
- **verify**: number of days that the user completed the deal but did not recieve any cash.
- **time_spent**: time spent in the app during thier sessions (this is an important note, only time during sessions was provided.
- **funded**: number of days that the user engaged or verified a deal in the app that was funded.
- **selffunded**: number of days that the user engaged or verified a deal in the app that was self-funded.
- **first_activity**: the first day the user engaged with the app in any way (i.e., anything other than "gap".
- **streak_activity**: the maximum number of days that the user engaged with the app in a row.
- **future_redemptions**: the number of redemptions cashed in the future-- what we are hoping to predict (Y).
